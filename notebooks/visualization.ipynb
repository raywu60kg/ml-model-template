# Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
%matplotlib inline
```

```python
data = pd.read_csv("../data/iris.csv")
data.head()
```

```python
def plot_data(data):
    # Visualize the data sets
    plt.figure(figsize=(16, 6))
    plt.subplot(1, 2, 1)
    names = np.unique(data['variety'])
    for target, target_name in enumerate(names):
        X_plot = data[data["variety"] == target_name].reset_index(drop=True)
        plt.plot(X_plot['sepal.length'], X_plot['sepal.width'], linestyle='none', marker='o', label=target_name)
    plt.xlabel('sepal.length')
    plt.ylabel('sepal.width')
    plt.axis('equal')
    plt.legend();

    plt.subplot(1, 2, 2)
    for target, target_name in enumerate(names):
        X_plot = data[data["variety"] == target_name].reset_index(drop=True)
        plt.plot(X_plot['petal.length'], X_plot['petal.width'], linestyle='none', marker='o', label=target_name)
    plt.xlabel('petal.length')
    plt.ylabel('petal.width')
    plt.axis('equal')
    plt.legend();
```

```python
plot_data(data)
```
